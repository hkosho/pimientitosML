```python
#-37  - BorderlineSMOTE -
# Ver.Multiclass Classification
# 
# -------------
# 
# This code is created by based below infomation.
# Qiita (SMOTE, ADASYN, Borderline-SMOTE, Safe-level SMOTE)
# @eigs 
# https://qiita.com/eigs/items/9ed36b5fc2913110e940
# 
# -------------

# 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 
Ndata = 3000;

# 
data  = np.random.rand(Ndata, 2)

# 
x = data[:, 0]    # X
y = data[:, 1]    # Y

# 
# True/False

# (x, y), ()
# () x0.4, y0.5, ()0.3
idx1a = np.power((x-0.4), 2) + np.power((y-0.5), 2) < np.power(0.3, 2)

# () x0.9, 0.9, ()0.05
idx1b = np.power((x-0.9), 2) + np.power((y-0.9), 2) < np.power(0.05, 2)

#  x0.8, 0.15, ()0.1
idx2  = np.power((x-0.8), 2) + np.power((y-0.15), 2) < np.power(0.1, 2)

# 4
undersampleRate = 4

# 
# 
data1  = data[idx1a|idx1b, :]            # True/FalseTrue
data1  = data1[0:-1:undersampleRate, :]  # 

# 
data2  = data[idx2, :]
data2  = data2[0:-1:undersampleRate, :]

# 
# True/False
data = np.delete(data, np.where(data[idx1a|idx1b|idx2, :]), axis=0)

# 
plt.figure(figsize=(7,7), dpi=100)

# 
plt.scatter( data[:, 0],  data[:, 1], c='white', marker='o', s=50, linewidths=2.0, edgecolors='gray',  alpha=0.3, label='class0:majority')
plt.scatter(data1[:, 0], data1[:, 1], c='white', marker='o', s=50, linewidths=2.0, edgecolors='blue',  alpha=0.9, label='class1:minority1(idx1a, idx1b)')
plt.scatter(data2[:, 0], data2[:, 1], c='white', marker='o', s=50, linewidths=2.0, edgecolors='green', alpha=0.9, label='class2:minority2(idx2)')

# 
plt.title('Sample Dataset')

# 
plt.legend(loc='upper left')

# 
plt.grid()

plt.show()
```

```python
# 
# 
import numpy as np
import numpy.matlib

# ()
dataset = np.vstack((data, data1, data2))

# 
label0 = np.matlib.repmat('0', len(data),  1)  # class0
label1 = np.matlib.repmat('1', len(data1), 1)  # class1()
label2 = np.matlib.repmat('2', len(data2), 1)  # class2()

# 
labels = np.vstack((label0, label1, label2))

# 
dataset = np.hstack((dataset, labels))
```

```python
# 

# 
from matplotlib.lines import Line2D  # 

# 
plt.figure(figsize=(7,7), dpi=100)

# ()
for element in dataset:

    # 
    if element[2] == '0':
        plt.scatter(float(element[0]), float(element[1]), c='white', marker='o', s=50, linewidths=2.0, edgecolors='cyan',  alpha=0.3)
        
    # 1(idx1a(), idx1b())
    elif element[2] == '1':
        plt.scatter(float(element[0]), float(element[1]), c='white', marker='o', s=50, linewidths=2.0, edgecolors='red',  alpha=0.9)
        
    # 2(idx2)
    elif element[2] == '2':
        plt.scatter(float(element[0]), float(element[1]), c='white', marker='o', s=50, linewidths=2.0, edgecolors='magenta', alpha=0.9)
       
    # 
    else:
        pass
    
# 
legend_elements = [Line2D([],[], marker='o', markersize=8, linewidth=2.0, mfc='white', mec='cyan',    lw=0.0, label='class0:majority'), \
                   Line2D([],[], marker='o', markersize=8, linewidth=2.0, mfc='white', mec='red',     lw=0.0, label='class1:minority1(idx1a, idx1b)'), \
                   Line2D([],[], marker='o', markersize=8, linewidth=2.0, mfc='white', mec='magenta', lw=0.0, label='class2:minority2(idx2)')]

# 
plt.title('Sample Dataset after join')

# 
plt.legend(handles=legend_elements, loc='upper left')

# 
plt.grid()

plt.show()
```

## sampling_strategy='not majority' 

```python
# BorderlineSMOTE(Ver.not majority)

# 
from imblearn.over_sampling import BorderlineSMOTE

sm = BorderlineSMOTE(sampling_strategy='not majority',\
                     random_state=None,\
                     k_neighbors=5,\
                     n_jobs=None,\
                     m_neighbors=10,\
                     kind='borderline-1')

X_res, y_res = sm.fit_resample(X, y)
```
